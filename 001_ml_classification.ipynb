```python
# Load Libraries
# Wrangling
import numpy as np
import pandas as pd
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
# ML
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
# Read Input File
df = pd.read_csv("creditcard.csv")
print(df.shape)
df.head()
```

We will visualize the variables now

```python
fig = plt.figure(figsize=(15, 12))

# Loop through V1 to V28 for the scatterplots
for i in range(1, 29):
    plt.subplot(5, 6, i)
    plt.scatter(
        range(len(df[f"V{i}"])),  # X-axis: indices
        df[f"V{i}"],  # Y-axis: column values
        c=df["Class"].map({0: "green", 1: "red"}),  # Color by class
        s=10  # Size of the circles
    )
    plt.title(f"V{i}", fontsize=8)

# Add the "Amount" column as the last subplot
plt.subplot(5, 6, 29)
plt.scatter(
    range(len(df["Amount"])),
    df["Amount"],
    c=df["Class"].map({0: "green", 1: "red"}),  # Color by class
    s=10  # Size of the circles
)
plt.title("Amount", fontsize=8)

plt.tight_layout()  # Adjust spacing for better visuals
plt.show()
```

```python
# Features and Target
X = df.drop(columns=["Class", "Time"])
y = df["Class"]
# Initialize model
logreg = LogisticRegression(solver="liblinear",random_state=42)
# Stratified k fold cross validation
n_splits = 5
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)
# Metrics
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []
# Perform Stratified K Fold Cross Validation
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    
    # Fit the model
    logreg.fit(X_train, y_train)
    
    # Predict on the test set
    y_pred = logreg.predict(X_test)
    
    # Compute metrics
    accuracy_list.append(accuracy_score(y_test, y_pred))
    precision_list.append(precision_score(y_test, y_pred))
    recall_list.append(recall_score(y_test, y_pred))
    f1_list.append(f1_score(y_test, y_pred))

# Display average metrics
print(f"Accuracy: {sum(accuracy_list) / n_splits:.4f}")
print(f"Precision: {sum(precision_list) / n_splits:.4f}")
print(f"Recall: {sum(recall_list) / n_splits:.4f}")
print(f"F1 Score: {sum(f1_list) / n_splits:.4f}")
```
